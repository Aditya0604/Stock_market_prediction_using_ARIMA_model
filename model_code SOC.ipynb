# Differencing
Both, from the graph and from the adfuller test we could see that the data is not stationary, as the p value is quite high, thus we need to make the data stationary first in order to train the model.

Thus to make the data stationary we try differencing (i.e. shift the data) as:-

```python
ICICI_data_1yr_diff = ICICI_data_1yr - ICICI_data_1yr.shift()
```

```python
result = adfuller(ICICI_data_1yr_diff.dropna(), autolag='AIC')
```

```python
print(f'ADF Statistic: {result[0]}')
print(f'n_lags: {result[1]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}') 
```

```python
plt.plot(ICICI_data_1yr_diff.dropna());
```

```python
rel_data_6m_diff = rel_data_6m - rel_data_6m.shift()
```

```python
result = adfuller(rel_data_6m_diff.dropna(), autolag='AIC')
```

```python
print(f'ADF Statistic: {result[0]}')
print(f'n_lags: {result[1]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}') 
```

Now, after applying auto arima model to our data we got (0,0,0) as the best fit parameters so lets try them out, and train our model on it to get the results.

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(ICICI_data_6m_diff.dropna(), order=(0,0,0))  
results_ARIMA = model.fit()
```

```python
ARIMA_diff_predictions = pd.Series(results_ARIMA.fittedvalues, copy=True)
print(ARIMA_diff_predictions)
```

As we differenced the datat , we need to take the commulative that will have the opposite effect of differencing

```python
ARIMA_diff_predictions_cumsum = ARIMA_diff_predictions.cumsum()
print(ARIMA_diff_predictions_cumsum.head())
```

Now, we get the commulative sum, we would add the first value of the series so that every corresponding value falls in its place

```python
ARIMA_log_prediction = pd.Series(ICICI_data_6m.iloc[0], index=ICICI_data_6m.index)
ARIMA_log_prediction = ARIMA_log_prediction.add(ARIMA_diff_predictions_cumsum,fill_value=0)
ARIMA_log_prediction.head()
```

Now, we got our fitted values, lets plot it to see what we have

```python
plt.figure(figsize=(12,8))
plt.plot(ICICI_data_6m)
plt.plot(ARIMA_log_prediction)
plt.legend(["original values","predicted values"])
```
